```python
import pandas as pd
import json
```

MATCH

```python
with open('t20_wc_match_results.json','r') as f:
  data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match = df_match.rename(columns={'scorecard':'match_id'})
df_match.head()
```

```python
match_dictionary = {}
for index,row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']
  match_dictionary[key1] = row['match_id']
  match_dictionary[key2] = row['match_id']
match_dictionary
```

BATTING

```python
with open('t20_wc_batting_summary.json','r') as f:
  data = json.load(f)
all_data = []
for rec in data:
  all_data.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_data)
df_batting
```

```python
df_batting["out_or_not_out"] = df_batting["dismissal"].apply(lambda x: 'not_out' if len(x) == 0 else 'out')
df_batting
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting
```

```python
df_batting['match_id'] = df_batting['match'].map(match_dictionary)
df_batting
```

BOWLING


```python
with open('t20_wc_bowling_summary.json','r') as f:
  data = json.load(f)
all_data = []
for rec in data:
  all_data.extend(rec['bowlingSummary'])
df_bowling = pd.DataFrame(all_data)
df_bowling
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_dictionary)
df_bowling
```

PLAYERS

```python
with open('t20_wc_player_info.json','r') as f:
  data = json.load(f)

df_players = pd.DataFrame(data)
df_players
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†',''))
df_players
```

Convert Dataframe to CSV

```python
df_match.to_csv('dim_match_summary.csv',index=False)
df_batting.to_csv('fact_batting_summary.csv',index=False)
df_bowling.to_csv('fact_bowling_summary.csv',index=False)
df_players.to_csv('dim_players_summary.csv',index=False)
```
